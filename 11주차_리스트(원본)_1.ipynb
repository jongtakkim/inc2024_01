```python
# 1분퀴즈 3
# 0 ~ 100까지 30개의 radnom 값을 score 리스트에 저장하고
# 30개의 값중에 80점 이상인 값은 A 리스트
# 30개의 값중에 40점 이상인 값은 B 리스트
# 30개의 값중에 40점 미만인 값은 C 리스트에 저장하고 출력하여라
import random

score = []
A = []
B = []
C = []

for i in range(30) :
    score.append(random.randint(0, 100))

for i in range(len(score)) :
    if score[i] >= 80 :
        A.append(score[i])
    elif score[i] >= 40 :
        B.append(score[i])
    else :
        C.append(score[i])
print(score)
print(A)
print(B)
print(C)
```

```python
#1번 - 주민번호 앞6자리와 뒷자리 1개를 입력받아 생년월일과 성별을 출력하세요
jumin = input('주민번호 앞 6잘와 뒷자리 1개 입력(XXXXXX-X) ->')
year = jumin[:2]
year = '20' + year
month = jumin[2:4]
day = jumin[4:6]

if jumin[-1:] == '3' :
    sex = '남자'
else :
    sex = '여자'

print(f'생년월일 : {year}년 {month}월 {day}일 성별은 {sex}입니다.')
```

```python
# 2번 - 수강과목수와 각각의 수강 과목점수를 입력받아 리스트에 저장하고
# 80점 이상인 과목은 “합격", 80점 미만은 ”불합격“이라고 한다. 합격" 과 "불합격"한 과목이 수를 <출력결과>와 같이 출력하시오
score=[]
p,f=0,0
cnt=int(input('수강 과목 수 : '))

for i in range(cnt) :
  s = int(input(f'{i+1}번째 점수 : '))
  score.append(s)

for i in range(cnt) :
  if score[i] > 80 :
    p += 1
  else :
    f += 1

print(f"합격한 과목수 : {p}")
print(f'불합격한 과목수 : {f}')
```
